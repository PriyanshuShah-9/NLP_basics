```python
sentences = [
    ["natural","language","processing","is","a","very","interesting","field","of","study"],
    ["machine","learning","models","learn","patterns","from","data","over","time","efficiently"],
    ["deep","learning","uses","neural","networks","for","complex","tasks","and","predictions"],
    ["artificial","intelligence","is","changing","the","world","in","many","different","ways"],
    ["data","science","combines","statistics","programming","and","domain","knowledge","together","effectively"],
    ["students","enjoy","learning","new","concepts","in","technology","and","innovation","daily"],
    ["python","is","a","popular","programming","language","for","data","analysis","tasks"],
    ["text","processing","is","important","in","natural","language","understanding","applications","today"],
    ["learning","algorithms","improve","their","performance","with","more","training","data","available"],
    ["big","data","tools","help","manage","and","process","large","datasets","quickly"],
    ["supervised","learning","requires","labeled","data","for","training","machine","learning","models"],
    ["unsupervised","learning","discovers","hidden","patterns","in","unlabeled","datasets","automatically","efficiently"],
    ["feature","selection","helps","reduce","dimensionality","and","improve","model","accuracy","significantly"],
    ["neural","networks","contain","layers","of","interconnected","nodes","called","artificial","neurons"],
    ["language","models","predict","the","next","word","based","on","context","information"],
    ["data","visualization","makes","it","easier","to","understand","complex","information","quickly"],
    ["optimization","algorithms","are","used","to","minimize","loss","functions","during","training"],
    ["regularization","techniques","prevent","overfitting","in","machine","learning","models","effectively","always"],
    ["computer","vision","allows","machines","to","interpret","and","analyze","visual","data"],
    ["ai","systems","are","becoming","more","powerful","and","widely","used","across","industries"]
]
```

```python
for i, sentence in enumerate(sentences):
    word_count = len(sentence)
    print(f"Sentence {i+1}: {word_count} words")
```

```python
#Word2Vec
from gensim.models import Word2Vec

model = Word2Vec(
    sentences,
    vector_size=50,
    window=3,
    min_count=1,
    workers=4,
)
```

```python
print("Vector for language")
print(model.wv['language'])

print("\nWords similar to language ")
print(model.wv.most_similar('language'))
```

```python
print("Vector for learning")
print(model.wv['learning'])

print("\nWords similar to learning ")
print(model.wv.most_similar('learning'))
```

```python
print("Vector for data")
print(model.wv['data'])

print("\nWords similar to data ")
print(model.wv.most_similar('data'))
```

```python
print(model.wv.similarity("data", "science"))
```
